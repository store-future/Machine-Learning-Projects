```python
import pandas as pd 
import ast                  # used to convert list string into list 
```

```python
movies = pd.read_csv('./datasets/tmdb_5000_movies.csv')
credits = pd.read_csv('./datasets/tmdb_5000_credits.csv')
```

```python
movies = movies.merge(credits , on='title')
```

```python
# selecting only required columns
required_col = ['movie_id' , 'title' , 'overview' , 'genres' , 'keywords' , 'cast' , 'crew']
movies = movies[required_col]
movies.head(2)
```

```python
# checking if movie dataframe has null value or not
movies.isnull().sum()
```

```python
# droping null values rows in dataframe
movies.dropna(inplace=True)
```

```python
# our genres item list is in string list so we have  to convert it by ast.literal_eval
movies.iloc[0].genres
```

```python
# converting each element of overview into listitems  
movies['overview'] = movies['overview'].apply(lambda x:x.split())
movies['overview'].iloc[0:2]
```

```python
movies['genres'].iloc[0]
```

```python
# function that is breaking genres elemets into list item
def convert_into_listitem(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
# converting each element of genres into seperatable list item
movies['genres'] = movies['genres'].apply(convert_into_listitem)
```

```python
movies['genres'][0]
```

```python
movies['keywords'] = movies['keywords'].apply(convert_into_listitem)
```

```python
movies['keywords'][0]
```

```python
#  function that is coverting cast elemets into list item
def convert_cast_listitem(obj):
    L = []
    count = 0
    for i in ast.literal_eval(obj):
        if count !=3:
           L.append(i['name'])
           count+=1
        else:
            break
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert_cast_listitem)
```

```python
movies['cast'][0]
```

```python
# funtion that is coverting crew elemets into list item where job = director
def convert_crew_listitem(obj):
    L = []

    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L
```

```python
movies['crew'] = movies['crew'].apply(convert_crew_listitem)
```

```python
movies['crew'][0]
```

```python
movies.head(2)
```

```python
# removing any space between full names from all columns  ex -science fiction =sciencefiction
movies['genres']   = movies['genres'].apply(lambda x:[i.replace(" " ,"") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" " ,"") for i in x])
movies['cast']     = movies['cast'].apply(lambda x:[i.replace(" " ,"") for i in x])
movies['crew']     = movies['crew'].apply(lambda x:[i.replace(" " ,"") for i in x])
```

```python
movies.head(2)
```

```python
# merging overview , genres ,keywords ,cast ,crew into one new column = tags 
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew'] 
```

```python
movies.head(2)
```

```python
# creating a new df with selected columns
new_df = movies[['movie_id' , 'title' , 'tags']]
```

```python
new_df.head(2)
```

```python
# converting list of each element of tags into string format
new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x))
```

```python
new_df['tags'][0]
```

```python
new_df.head()
```

```python
new_df.shape
```

```python
# lowering case the elements of tags series
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

new_df['tags'][0]

 # Till here we do Data Preprocessing , now we apply Machine Learning algorithems  


```python
# converting tag series element into base root form of english
from nltk.stem.porter import PorterStemmer
ps =PorterStemmer()
```

```python
def stemming_engine(obj):
    L = []
    for i in obj.split():
        L.append(ps.stem(i))
    return " ".join(L)
```

```python
new_df['tags'] = new_df['tags'].apply(stemming_engine)
```

```python
new_df['tags'][0]
```

```python
The CountVectorizer is a text feature extraction technique. It  converts them into a numerical representation

from sklearn.feature_extraction.text import CountVectorizer
algo = CountVectorizer(max_features=5000 , stop_words='english')
```

```python
#first it will learn those 5000 words then make vectors 
# use .toarray to convert sparse matrix into array

vectors = algo.fit_transform(new_df['tags']).toarray()
```
